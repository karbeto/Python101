```python
import pandas as pd
import numpy as np
from pathlib import Path
```

#### Filtering with Pandas


```python
path = Path('.', 'data', 'SuperStore.xlsx')
df = pd.read_excel(path, sheet_name='Orders')
df.head()
pd.DataFrame(df.dtypes, columns=['type'])
```

```python
df['Customer Name'].unique()
```

```python
#In how many Invoices (Main Orders), were the orders placed?
customer_name = 'Claire Gute'

df[  df['Customer Name'] == customer_name   ]['Order ID'].nunique()
```

```python
#How many orders were placed in the 'Consumer' segment?

df[  df['Segment'] ==  'Consumer' ]['Order ID'].nunique()
```

```python
#Which customers made purchases in 'Consumer' segment

customer = set(df[df['Segment'] == 'Consumer']['Customer Name'])

customer
```

```python
#Filter Sales greater then 800
df[df['Sales'] > 800].head()
```

```python
#Chaining filters, to get sales between 800 and 1000
df[(df['Sales'] > 800) & (df['Sales'] < 1000) ].head()
```

```python
# Filter Sales whose product name length is < 50
df[df['Product Name'].str.len() < 50].head()
```

```python
# Filter Sales between 700 and 900, in state Florida or Kentucky (solution 2 - more elegant and more readable!)
states = ['Florida', 'Kentucky']

df[(df['Sales'] > 700) & (df['Sales'] < 900) & (df['State'].isin(states))].head()
```

```python
# Filter Sales outside of the states Florida and Kentucky
df[~df['State'].isin(['Florida', 'Kentucky'])].head()
```

```python
#Filter sales that are in December

df[df['Order Date'].dt.month == 12].head()
```

#### Query Method

```python
# Filter Sales outside of the states Florida and Kentucky
df.query("State not in ['Florida', 'Kentucky']").head()
```

```python
df.query("Sales > 100 and Sales < 800 and Category == 'Technology'").head()
```

```python
#If column has " " in the title we need to use backticks ``
df.query("`Product Name`.str.startswith('B')").head()
```

```python
df.query("`Product Name`.str.upper().str.startswith('B')").head()
```

#### Projection(filtering with pandas)

```python
#Select the customers whose name starts with 'Ad' and show only the Customer Name, the State and the Sales for these customers
df[ df['Customer Name'].str.startswith('Ad') ][['Customer Name', 'State', 'Sales']]
```

```python
#Select the customers whose name does not start with 'A' and show only the Customer Name, the State and the Sales for these customers.
#Sort the result Customer Name ASC, Sales DESC
df[ ~df['Customer Name'].str.startswith('A')].sort_values(by=['Customer Name', 'Sales'], ascending=[True, False])
```

```python
#Select the Customer name, Category, Sales and Profit, for all sales over 1500 dollars, that happened in Houston

df[(df['City'] == 'Houston') & (df['Sales'] > 1500)][['Customer Name', 'Category', 'Sales', 'Profit']]
```

```python
df.nlargest(5, 'Profit')
```
